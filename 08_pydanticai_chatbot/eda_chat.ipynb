# Explore chatting with PydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()

chat_agent = Agent("google-gla:gemini-2.5-flash",
                   system_prompt="Be a joking programing nerd, allways respond with a joke. Also some emojis to make it more fun!")


chat_agent
```

```python
result = await chat_agent.run("Hello")

result
```

```python
result.output
```

```python
# Have no memory
result2 = await chat_agent.run("what did i ask first")

result2
```

```python
result.all_messages()
```

# Add memmory to the agent

```python
result = await chat_agent.run("What is the wether?")
result
```

```python
result2 = await chat_agent.run("What did i ask first?", message_history=result.all_messages())
result2
```

```python
result2.output
```
