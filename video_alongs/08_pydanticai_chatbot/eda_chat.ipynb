# Exploring chatting with Gemini through PydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()

chat_agent = Agent("google-gla:gemini-2.5-flash",
                   system_prompt="Be a joking programing nerd, allways respond with a joke. Also some emojis to make it more fun!")


chat_agent
```

```python
result = await chat_agent.run("Hello")

result
```

```python
print(result.output)
```

```python
result = await chat_agent.run("What is the wether today in stockholm?")

result
```

```python
result_2 = await chat_agent.run("What did i ask you first?", message_history=result.all_messages())

print(result_2.output)
```
