# Gemeni AIP intro

```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`/ GOOGLE_API_KEY.
client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash", 
    contents="Explain how AI works in a few words"
)
print(response.text)
```

```python
def ask_gemini(promt, model="gemini-2.5-flash"):
    response = client.models.generate_content(
        model=model, 
        contents=promt,
    )

    return response

response = ask_gemini("Give me 2 short jokes about programming.")
print(response.text)
```

## Thinking is expencive

```python
from google.genai import types

response = client.models.generate_content(
    model="gemini-2.5-flash", 
    contents="Give me 2 short jokes about programming.",
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(
            thinking_budget=0),
            )
    )

print(response.text)
```

## System Instruction

```python
system_instruction = """
You are an expert Python programmer and data engineer.
You are critical and do not accept any input that is not well defined.
Be conservative with your answers and do not make up facts.
"""

promt = """ 
Explain OOP in Python with an example.
"""


response = client.models.generate_content(
    model="gemini-2.5-flash", 
    contents=promt,
    config=types.GenerateContentConfig(
        system_instruction=system_instruction
    ),
)

print(response.text)
```
